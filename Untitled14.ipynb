Preparation Data

```python
x = df.drop(columns=['Hasil'])
y = df['Hasil']

print('x :', x.shape)
print('y :', y.shape)
```

```python
# standarisasi
# Inisialisasi objek StandardScaler
scaler = StandardScaler()
# StandardScaler akan mengubah data sehingga memiliki rata-rata (mean) 0 dan standar deviasi 1

# Menyesuaikan (fit) scaler dengan data 'x' (mencari mean dan standar deviasi dari 'x')
scaler.fit(x)

# Transformasi data 'x' agar memiliki distribusi dengan mean 0 dan standar deviasi 1
x = scaler.transform(x)
```

```python
# Spliting Data
# Membagi dataset menjadi data latih (training) dan data uji (testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

train_test_split(x, y, test_size=0.2, random_state=42)
Fungsi ini membagi dataset menjadi data latih dan data uji secara acak.

x â†’ Fitur (variabel independen) yang digunakan sebagai input model.
y â†’ Label atau target (variabel dependen) yang ingin diprediksi.
test_size=0.2 â†’ 20% dari data akan digunakan sebagai data uji, sementara 80% lainnya untuk data latih.
random_state=42 â†’ Menetapkan seed acak, sehingga setiap kali kode dijalankan, hasil pembagian data akan tetap sama (reproducible).

70% â†’ Training
15% â†’ Validation (untuk tuning hyperparameter)
15% â†’ Testing (untuk evaluasi akhir)

 Modeling dan Evaluation

```python
# Membuat model SVM dengan kernel linear
clf = SVC(kernel='linear')
# 'kernel=linear' berarti kita menggunakan SVM dengan pemisahan garis lurus

# Melatih model SVM dengan data latih (x_train dan y_train)
clf.fit(x_train, y_train)

# Menggunakan model SVM yang sudah dilatih untuk memprediksi data uji (x_test)
y_pred = clf.predict(x_test)

# Menghitung akurasi model dengan membandingkan prediksi (y_pred) dan label sebenarnya (y_test)
CLF_acc = accuracy_score(y_pred, y_test)

# Menampilkan laporan klasifikasi yang mencakup precision, recall, f1-score, dan support untuk setiap kelas
print(classification_report(y_test, y_pred))

# Menampilkan akurasi model dalam bentuk persentase dengan dua angka di belakang koma
print('Akurasi SVM :{:.2f}%'.format(CLF_acc * 100))
```

Testing

```python
new_data = {'Kehamilan': [1],
            'Glukosa':[85],
            'Tekanan Darah':[66],
            'Ketebalan Kulit':[29],
            'Insulin': [0],
            'BMI':[26.6],
            'DiabetesPedigreeFunction':[0.351],
            'Umur': [31]}
new_data = pd.DataFrame(new_data)
new_data
```

```python
# Melakukan standarisasi pada data baru menggunakan scaler yang sudah dilatih sebelumnya
scaled_new_data = scaler.transform(new_data)
# 'new_data' adalah data baru yang ingin diprediksi
# 'scaler.transform()' memastikan bahwa data baru memiliki skala yang sama dengan data latih
# Ini penting agar model dapat bekerja dengan baik, karena model dilatih dengan data yang sudah distandarisasi

# Menggunakan model SVM yang sudah dilatih untuk memprediksi kelas dari data baru
y_pred_new = clf.predict(scaled_new_data)
# Model akan mengklasifikasikan data baru berdasarkan pola yang telah dipelajari selama training

# Menampilkan hasil prediksi dari data baru
print('Diagnosis Data Baru :', y_pred_new)
# Hasil prediksi akan berupa label kelas, misalnya "0" atau "1" tergantung pada dataset yang digunakan
```
